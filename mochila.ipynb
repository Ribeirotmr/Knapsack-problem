```python
from qiskit_optimization.problems import QuadraticProgram
from qiskit_optimization.converters import QuadraticProgramToQubo
from qiskit_algorithms.minimum_eigensolvers import QAOA
from qiskit_algorithms.optimizers import COBYLA
from qiskit_aer.primitives import Sampler
from qiskit_algorithms.utils import algorithm_globals
import numpy as np
import matplotlib.pyplot as plt

# ---------- Configuração do problema ----------
N = 5
values = [4, 2, 2, 1, 10]      # Valores dos itens
weights = [12, 2, 1, 1, 4]     # Pesos dos itens
W = 15                        # Capacidade máxima da mochila
a = 50                        # Penalidade para violação da restrição

algorithm_globals.random_seed = 42


qp = QuadraticProgram()
for j in range(N):
    qp.binary_var(f'x_{j}')


objective = {f'x_{j}': values[j] for j in range(N)}
qp.maximize(linear=objective)

constraint = {f'x_{j}': weights[j] for j in range(N)}
qp.linear_constraint(linear=constraint, sense='LE', rhs=W, name='weight')


converter = QuadraticProgramToQubo(penalty=a)
qubo = converter.convert(qp)


optimizer = COBYLA()
sampler = Sampler()
qaoa = QAOA(sampler=sampler, optimizer=optimizer, reps=3)
result = qaoa.compute_minimum_eigenvalue(qubo.to_ising()[0])


bound_circuit = qaoa.ansatz.assign_parameters(result.optimal_parameters)
job = sampler.run([bound_circuit], shots=1024)
counts = job.result().quasi_dists[0].binary_probabilities()

states = list(counts.keys())
probabilities = list(counts.values())

plt.bar(range(len(states)), probabilities, tick_label=states)
plt.xlabel('Solução (x_4 x_3 x_2 x_1 x_0)')
plt.ylabel('Probabilidade')
plt.title('Distribuição de Probabilidades - QAOA para Mochila')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


viaveis = []
for state, prob in counts.items():
    x = [int(bit) for bit in state[::-1]]
    peso = sum(w * xi for w, xi in zip(weights, x))
    if peso <= W:
        valor = sum(v * xi for v, xi in zip(values, x))
        viaveis.append((state, prob, valor, peso))


if viaveis:
    melhor = max(viaveis, key=lambda t: t[2])
    melhor_x = [int(b) for b in melhor[0][::-1]]
    print(f'\nMelhor solução viável: x = {melhor_x}')
    print(f'Valor total = ${melhor[2]}, Peso total = {melhor[3]} kg')
    print(f'Probabilidade = {melhor[1]:.3f}')
else:
    print("Nenhuma solução viável encontrada.")
```

```python
import matplotlib.pyplot as plt


job = sampler.run([bound_circuit], shots=2048)
counts = job.result().quasi_dists[0].binary_probabilities()


solutions = []
for bitstring, prob in counts.items():

    x = [int(bit) for bit in bitstring[::-1]]
    total_value = sum(v * xi for v, xi in zip(values, x))
    total_weight = sum(w * xi for w, xi in zip(weights, x))
    is_valid = total_weight <= W
    solutions.append({
        'bitstring': bitstring,
        'x': x,
        'value': total_value,
        'weight': total_weight,
        'prob': prob,
        'valid': is_valid
    })


solutions = sorted(solutions, key=lambda d: d['prob'], reverse=True)


labels = [s['bitstring'] for s in solutions]
probs = [s['prob'] for s in solutions]
colors = ['green' if s['valid'] else 'red' for s in solutions]


plt.figure(figsize=(12, 6))
plt.bar(labels, probs, color=colors)
plt.xlabel('Solução (bitstring)')
plt.ylabel('Probabilidade')
plt.title('Distribuição de Probabilidades - QAOA para Mochila')
plt.xticks(rotation=90)
plt.grid(axis='y')


import matplotlib.patches as mpatches
green_patch = mpatches.Patch(color='green', label='Viável')
red_patch = mpatches.Patch(color='red', label='Inviável')
plt.legend(handles=[green_patch, red_patch])

plt.tight_layout()
plt.show()

print("\n Soluções:")
for s in solutions:
    status = "[SIM] Viável" if s['valid'] else "[NÃO] Inviável"
    print(f"{s['bitstring']} → x={s['x']}, valor={s['value']}, peso={s['weight']} → {status}, prob={s['prob']:.4f}")
```
